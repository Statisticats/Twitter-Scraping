# **Praproses Data**

```python
tweet.columns = ['user_tweeted', 'user_replied', 'verified']
tweet.head()
```

```python
# Drop NaN : Tweet that has no reply
tweet = tweet.dropna(axis=0)
tweet.head()
```

```python
tweet['pair_reply'] = tweet[['user_tweeted', 'user_replied']].apply(lambda x: ', '.join(x[x.notnull()]), axis = 1)
tweet.head()
```

```python
# Get Unique Pair
unique_pair = tweet.iloc[:,[3]]
unique_pair.head()
```

```python
#Count for weight
countreply = unique_pair.groupby('pair_reply', as_index=False).size()
countreply.head()
```

```python
# Delete Duplicate from rawdata
tweet = tweet.drop_duplicates()
tweet.head()
```

```python
#Merge Table
mydata = pd.merge(left=tweet, right=countreply, how='left', left_on='pair_reply', right_on='pair_reply')
mydata.head()
```

```python
#Get whose mentioned themeself
selfmentioned = mydata.loc[mydata['user_tweeted'] == mydata['user_replied']]
selfmentioned.head()
```

```python
#Get whose not mentioned themeself
notselfreplied = mydata.loc[mydata['user_tweeted'] != mydata['user_replied']]
notselfreplied.head()
```

```python
notselfreplied.sort_values(by='verified', ascending = False).head()
```

```python
# Get Verified Account Only
verified = notselfreplied.loc[notselfreplied['verified']==True]
verified.head()
```

```python
verified.to_csv('verified_account_with_reply.csv', index=False)
```

```python
# Get not Verified Account
unverified = notselfreplied.loc[notselfreplied['verified']==False]
unverified.head()
```

```python
unverified.to_csv('unverified_account_with_reply.csv', index=False)
```

# **Verified Account Network**

```python
verified.rename(columns = {'size':'weight'}, inplace = True)
```

```python
verified_data = verified.iloc[:,[0,1,4]]
verified_data.head()
```

```python
verified_data=pd.DataFrame(data=verified_data)
verified_data.head()
```

```python
verified_data.columns = ['source','target','weight']
verified_data.head()
```

```python
verified_data.sort_values(by=['weight'], ascending=False).head()
```

```python
#Most Replied User - verified
mostreplied = verified_data.iloc[:,[0,2]]
mostreplied = mostreplied.groupby('source', as_index=False).count()
mostreplied = mostreplied.sort_values(by=['weight'], ascending=False)
mostreplied.head()
```

```python
top10mostreplied_ver = mostreplied.head(10)
top10mostreplied_ver.rename(columns = {'weight':'replied_count'}, inplace = True)
top10mostreplied_ver
```

```python
# Plot most replied user
 
# Figure Size
plt.figure(figsize=(8, 6))
 
# Bar Plot
sns.barplot(x="source", y="replied_count", data=top10mostreplied_ver,
            palette='Set2')
 
# Add Plot Title
plt.title('Verified User yang paling banyak me-reply',
             loc ='center', fontsize=18)

plt.xticks(rotation=45)
 
plt.tight_layout()

plt.savefig('ver_most_replied_tweet.png')
# Show Plot
plt.show()
```

**Verified Account Network** 

```python
V1 = nx.from_pandas_edgelist(verified_data, 'source', 'target', edge_attr ='weight',create_using=nx.DiGraph())
```

```python
plt.figure(figsize = (12,10))

node_color = [V1.out_degree(v) for v in V1]
node_size = [200*V1.degree(v) for v in V1]
edge_width = [0.7*V1[u][v]['weight'] for u, v in V1.edges()]

graph_ver = nx.kamada_kawai_layout(V1)
nx.draw_networkx(V1, graph_ver, node_color = node_color, node_size = node_size,
                 with_labels=False, width=edge_width, edge_color = 'black', alpha = 0.9)

nx.draw_networkx_labels(V1, graph_ver, labels={'KAI121':'KAI121', 'smartfrenworld':'smartfrenworld'}, font_size=14, font_color='black')

plt.tight_layout()
plt.savefig('ver_all_out.png')
```

Pada network di atas terlihat ukuran node berbeda satu sama lain, hal ini disesuaikan dengan jumlah panah yang keluar dari node tersebut, dengan kata lain akun yang banyak membalas tweet akan memiliki ukuran yang lebih besar. Diketahui bahwa akun KAI121 dan smartfrenworld memiliki node paling besar. Hal ini berarti kedua akun tersebut banyak melakukan balasan terhadap tweet akun lain. Node yang ditunjuk oleh panah merupakan akun yang membagikan tweet, sedangkan node tempat garis panah dimulai merupakan user yang membalas. Jika diperhatikan terlihat ketebalan garis yang menghubungkan node terdapat perbedaan, ada yang lebih tebal dibandingkan yang lainnya. Ketebalan garis (edge) ini disesuaikan dengan bobot yang dimiliki edge. Semakin tebal garis menandakan user tersebut melakukan banyak balasan terhadap tweet yang dibagikan user lainnya. Banyaknya garis yang keluar dari node dapat disebut juga dengan out degree.

**Degree Verified Account Network**

```python
#Degree
degrees = [(node,val) for (node, val) in V1.out_degree()]
ver_degree=sorted(degrees, key=lambda x: x[1], reverse=True)
```

```python
ver_degree_df = pd.DataFrame(ver_degree)
ver_degree_df.columns = ['source', 'degree']
ver_degree_df.head()
```

```python
ver_degree_df_top10 = ver_degree_df.head(10)
```

```python
# Plot Top 10 Highest Degree

# Figure Size
plt.figure(figsize=(8, 6))
 
# Bar Plot
sns.barplot(x="degree", y="source", data=ver_degree_df_top10,
            palette='husl')
 
# Add Plot Title
plt.title("Top 10 Verified User's Out Degree",
             loc ='center', fontsize=18)
# Add Label
plt.xlabel('Jumlah Tweet')
plt.ylabel('User Name')

plt.tight_layout()
plt.savefig('top10_user_degree_ver.png')
# Show Plot
plt.show()
```

**Out-Degree Centrality**

*Out-Degree Centrality* merupakan salah satu cara untuk mengetahui node-node mana yang sangat penting dalam suatu network. Dalam *out-degree centrality*, node yang penting merupakan node yang memiliki *out-degree* yang tinggi. 

```python
# Degree Centrality
outdeg_cent = nx.out_degree_centrality(V1)
```

```python
out_cent_items = outdeg_cent.items()
out_cent_list = list(out_cent_items)
out_cent_df = pd.DataFrame(out_cent_list)
```

```python
out_cent_df.columns = ['target','outdeg_cent']
out_cent_df.head()
```

```python
out_cent_df.sort_values(by=['outdeg_cent'], ascending=False).head()
```

Berdasarkan *out-degree centrality*, akun KAI121 merupakan akun dengan nilai centrality tertinggi yaitu 0,207 sehingga akun KAI121 dinyatakan sebagai akun yang paling penting dalam network ini. Setelah dipelajari lebih jauh akun KAI121 melakukan balasan terhadap tweet akun lain yang bertanya mengenai aturan atau hal-hal terkait yang harus diperhatikan melakukan perjalanan menggunakan kereta api.

# Unverified Account Network

```python
unverified = pd.read_csv('unverified_account_with_reply.csv')
unverified.head()
```

```python
unverified.rename(columns = {'size':'weight'}, inplace = True)
unverified.head()
```

```python
unverified_data = unverified.iloc[:,[0,1,4]]
unverified_data.head()
```

```python
unverified_data=pd.DataFrame(data=unverified_data)
unverified_data.columns = ['source','target','weight']
unverified_data.head()
```

**Degree Unverified Account Network**

```python
U = nx.from_pandas_edgelist(unverified_data, source='source', target='target', edge_attr='weight', create_using=nx.DiGraph())
```

```python
#Degree
degrees_unver = [(node,val) for (node, val) in U.out_degree()]
unver_degree=sorted(degrees_unver, key=lambda x: x[1], reverse=True)
```

```python
unver_degree_df = pd.DataFrame(unver_degree)
unver_degree_df.columns = ['nodes', 'degree']
unver_degree_df.head()
```

```python
unver_degree_top10 = unver_degree_df.head(10)
```

```python
unver_degree_df_top = unver_degree_df.sort_values(by=['degree'], ascending=False).head(10)
#unver_degree_df_top = unver_degree_df_top.iloc[:,[0]]
unver_degree_df_top.head()
```

```python
unver_degree_df_top = unver_degree_df.sort_values(by=['degree'], ascending=False).head(10)
unver_degree_df_top = unver_degree_df_top.iloc[:,[0]]
unver_degree_df_top = pd.merge(left=unver_degree_df_top, right=unverified_data, how='left', left_on='nodes', right_on='source')
unver_degree_df_top
```

```python
# Plot Top 10 Highest Degree

# Figure Size
plt.figure(figsize=(8, 6))
 
# Bar Plot
sns.barplot(x="degree", y="nodes", data=unver_degree_top10,
            palette='husl')
 
# Add Plot Title
plt.title("Top 10 Unverified User's Out Degree",
             loc ='center', fontsize=18)
# Add Label
plt.xlabel('Jumlah Tweet')
plt.ylabel('User Name')

plt.tight_layout()
plt.savefig('top10_user_degree_unver.png')
# Show Plot
plt.show()
```

**Unverified Account Network**

Unverified User yang melakukan balasan (reply) berjumlah sangat besar yaitu 56,508 unique user (node) yang membentuk 48,974 edge. Dikarenakan ukuran data yang sangat besar ini, package networkx dari python tidak mampu menampilkan keseluruhan network sehingga pada bagian ini akan ditampilkan network dari 10 user yang terbanyak membalas tweet.

```python
# Create Graph
U1 = nx.from_pandas_edgelist(unver_degree_df_top, 'source', 'target', edge_attr ='weight',create_using=nx.DiGraph())

plt.figure(figsize = (12,10))

node_color = [U1.out_degree(v) for v in U1]
node_size = [100*U1.degree(v) for v in U1]
edge_width = [0.1*U1[u][v]['weight'] for u, v in U1.edges()]

graph_ver = nx.kamada_kawai_layout(U1)
nx.draw_networkx(U1, graph_ver, node_color = node_color, node_size = node_size,
                 with_labels=False, width=edge_width, edge_color = 'black', alpha = 0.9)

# Add Labels
nx.draw_networkx_labels(U1, graph_ver, labels={'dikarunia':'dikarunia', 
                                               'anggarasamudr':'anggarasamudr',
                                              'ysbunazgoef':'ysbunazgoef',
                                              'caewonzr':'caewonzr',
                                              'koobobaa':'koobobaa',
                                              'faktaidn_':'faktaidn_',
                                              'melodiysore':'melodiysore',
                                              '_bluecooll':'_bluecooll',
                                              'pxtriming':'pxtriming', 
                                              'UnfaedahB':'UnfaedahB'}, font_size=14, font_color='black')

plt.tight_layout()
plt.savefig('unver_network.png')
```

Terlihat ada 3 akun yang paling banyak membalas tweet yaitu dikarunia, anggarasamudr, dan ysbunazgoef. Setelah ditelusuri, akun @dikarunia merupakan akun yang membagikan kode promo untuk penginapan.
